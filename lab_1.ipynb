```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class_0 = np.random.normal( loc=15, scale=2, size=[50,2])
class_1 = np.random.normal( loc=0, scale=2, size=[70,2])
```

```python
class_0[0]
```

```python
labels = np.array([-1]*len(class_0) + [1]*len(class_1))
len(labels)
```

```python
flipp = np.random.randint(low=0, high=120, size=10)
print(flipp)
```

```python
print(labels[flipp])
```

```python
labels[flipp] = -1*labels[flipp]
```

```python
print(labels[flipp])
```

```python
plt.scatter(class_0[:,0], class_0[:,1], marker = 'o', c=labels[:50])
plt.scatter(class_1[:,0], class_1[:,1], marker = 'x', c=labels[50:])

plt.show()
```

```python
theta = np.random.uniform(1.5*np.pi, 4.5*np.pi, 1000)
phi = np.random.uniform(0,10,1000)
```

```python
x = theta * np.cos(theta)
y = phi * np.sin(phi)
z = phi
```

```python
plt.scatter(x,y)
```

```python
%matplotlib inline
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(x,y,z)
```

```python
data = {'x': x, 'y': y, 'z': z}
```

```python
df = pd.DataFrame(data)
df.head()
```
